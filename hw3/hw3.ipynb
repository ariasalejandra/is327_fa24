```python
import pandas as pd
from sklearn import tree, metrics

# Read in dataset
df = pd.read_csv('dataset.csv')

# Subset training data
train_df = df.iloc[:200]

# Subset testing data
test_df = df.iloc[200:]

# Train model and test
model = tree.DecisionTreeClassifier(min_samples_leaf=5)
features = ["col0", "col1", "col2", "col3", "col4", "col5", "col6", "col7", "col8", "col9", "col10", "col11", "col12", "col13", "col14"]
train_X = train_df[features]
test_X = test_df[features]
model.fit(train_X, train_df.label)
predictions = model.predict(test_X)

# Calculate accuracy (Cohen's kappa, in this case)
acc = metrics.cohen_kappa_score(test_df.label, predictions)
```
