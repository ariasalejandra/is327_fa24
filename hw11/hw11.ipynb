```python
import pandas as pd
from sklearn import preprocessing, cluster, metrics
from sklearn.metrics import silhouette_samples, silhouette_score

# Read in dataset
df = pd.read_csv('dataset-clustering.csv')
features = ['x', 'y']
# Preprocess the data with StandardScaler
scaler = preprocessing.StandardScaler()
# Create the model
scaled = scaler.fit_transform(df)
model = cluster.KMeans(2)
model.fit(scaled)
clusters_ids = model.predict(scaled)

# Find clusters in the data

# Calculate silhouette scores for the clusters found
silhouette_scores = silhouette_samples(scaled, clusters_ids)
mean_silhouette = silhouette_score(scaled, clusters_ids)
print("Silhouette Scores for each instance:", silhouette_scores)
print("Mean Silhouette Score:", mean_silhouette)
```
